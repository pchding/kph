```python
import numpy as np
import pandas as pd
```

```python
datao = pd.read_pickle("~/OneDrive/kph/processed2.pkl")
```

```python
datao['extraction_rate'] = datao['ext perc']/(datao['keywords'].apply(lambda x: len(x)))
```

```python
datao['no_key_word'] = datao['keywords'].apply(lambda x: len(x))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="darkgrid")
ax = sns.countplot(x="ext perc", data=datao)
```

```python
x = datao['ext perc']
sns.kdeplot(x, shade=True);
sns.distplot(x);
```

```python
sns.distplot(datao['extraction_rate'], kde=False, color='red', bins=20)
plt.title('number of extracted key phases', fontsize=14)
plt.xlabel('extraction rate', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
```

```python
dp = sns.distplot(datao['no_key_word'], kde=False, color='red', bins=200)
dp.set(xlim=(0, 30))
plt.title('number of key words', fontsize=14)
plt.xlabel('key word', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
```
